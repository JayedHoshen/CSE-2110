```python
#=========import section==========
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('data.csv')
```

```python
# ======================>>> Some Analysis of full Data sheet <<<====================
```

```python
#<<1>>---------- Provides basic information about the dataframe ----------
df.info()
```

```python
# <<2>>---------- Count Weather Frequency ------------
loc = df['Weather'].value_counts()
dfWeather = pd.DataFrame(loc)
dfWeather = dfWeather.reset_index()
dfWeather.columns = ['Weather', 'Frequency']

dfWeather
```

```python
#<<3>>---------- Display a bar-chart of showing of weather conditions ----------
keep = dfWeather[dfWeather['Frequency'] > 100]

color = ['#123524', '#3E7B27', '#EFE3C2', '#85A947', '#D39D55']
plt.subplots(figsize = (14, 6))
plt.bar(keep['Weather'], keep['Frequency'], color = color, width = 0.6, edgecolor = 'black', linewidth = 1, alpha = 0.6)
plt.xlabel('Weather')
plt.ylabel('Frequency')
plt.title("Showing Weather Conditions")
for i, v in enumerate(keep['Frequency']):
    plt.text(i, v, str(v), color = 'orange', ha = 'center', va = 'bottom')
plt.show()
```

```python
#<<8>>--------- Find the number of times when the "Minimum, Maximum and Average Wind Speed" --------
mn = df['Wind Speed_km/h'].min()
mx = df['Wind Speed_km/h'].max()
avg = df['Wind Speed_km/h'].mean()
print("---- Wind Speed km/h -----\n")
print("Minimum = ", mn)
print("Maximum = ", mx)
print("Average = ", avg)
```

```python
#<<9>>--------- Find the number of times when the "Minimum, Maximum and Average Visibility_km" --------
mn = df['Visibility_km'].min()
mx = df['Visibility_km'].max()
avg = df['Visibility_km'].mean()
print("----- Visibility km ----\n")
print("Minimum = ", mn)
print("Maximum = ", mx)
print("Average = ", avg)

# visibility greater than average
print("\n---------------- Visibility greater than average value -----------")
df[df['Visibility_km'] > avg]

#--------- Line chart of Date/Time and Visibility in this year -----------
# plt.subplots(figsize = (15, 6))
# plt.plot(df['Date/Time'], df['Visibility_km'])
# plt.title('Visibility (km) in this Year')
# plt.xlabel('Date/Time')
# plt.ylabel('Visibility_km')
# plt.show()
```

```python
#<<10>>--------- Find the number of times when the "Minimum, Maximum and Average Temperature" --------
mn = df['Temp_C'].min()
mx = df['Temp_C'].max()
avg = df['Temp_C'].mean()
print("----- Temperature ----\n")
print("Minimum = ", mn)
print("Maximum = ", mx)
print("Average = ", avg)

#--------- Line chart of Date/Time and Temperature in this year -----------
plt.subplots(figsize = (15, 6))
plt.plot(df['Date/Time'], df['Temp_C'])
plt.title('Temperature in 2012')
plt.xlabel('Date/Time')
plt.ylabel('Temperature')
plt.show()
```

```python
#<<11>>--------- Find the number of times when the "Minimum, Maximum and Average Press_kPa" --------
mn = df['Press_kPa'].min()
mx = df['Press_kPa'].max()
avg = df['Press_kPa'].mean()
print("----- Pressure kPa ----\n")
print("Minimum = ", mn)
print("Maximum = ", mx)
print("Average = ", avg)

# Pressure greater than average
print("\n---------------- Pressure greater than average value -----------")
df[df['Press_kPa'] > avg]
```

```python
# --------------- Showing a pie chart Where pressure is greater than 50 kPa -----------
y = df['Press_kPa'].value_counts()
dp = pd.DataFrame(y)
dp = dp.reset_index()
dp.columns = ['Press_kPa', 'Frequency']
py = dp[dp['Frequency'] > 50]
plt.pie(py['Frequency'], labels = py['Press_kPa'], textprops = {"fontsize" : 12}, autopct = "%0.2f%%", radius = 1.5)

plt.show()
```

```python
#<<12>>--------- Find the number of times when the "Minimum, Maximum and Average Dew Point Temp_C" --------
mn = df['Dew Point Temp_C'].min()
mx = df['Dew Point Temp_C'].max()
avg = df['Dew Point Temp_C'].mean()
print("----- Dew Point Temperature ----\n")
print("Minimum = ", mn)
print("Maximum = ", mx)
print("Average = ", avg)

# Dew Point Temp_C greater than average
print("\n---------------- Dew Point Temperature greater than average value -----------")
df[df['Dew Point Temp_C'] > avg]
```
