```python
import pandas as pd
import numpy as np
```

```python
s=pd.Series([1,2,3,4],
            index=["a","b","c","d"])
s
```

```python
s["a"]
```

```python
s2=s.reindex(["b","d","a","c","e"])
s2
```

```python
s3=pd.Series(["blue","yellow","purple"],
             index=[0,2,4])
s3
```

```python
s3.reindex(range(6),method="ffill")
```

```python
df=pd.DataFrame(np.arange(9).reshape(3,3),
                index=["a","c","d"],
                columns=["Tim","Tom","Kate"])
df
```

```python
df2=df.reindex(["d","c","b","a"])
df2
```

```python
names=["Kate","Tim","Tom"]
df.reindex(columns=names)
```

```python
df.loc[["c","d","a"]]
```

```python
s=pd.Series(np.arange(5.),
            index=["a","b","c","d","e"])
s
```

```python
new_s=s.drop("b")
new_s
```

```python
s.drop(["c","d"])
```

```python
data=pd.DataFrame(np.arange(16).reshape(4,4),
                  index=["Kate","Tim",
                         "Tom","Alex"],
                  columns=list("ABCD"))
data
```

```python
data.drop(["Kate","Tim"])
```

```python
data.drop("B",axis=1)
```

```python
data.drop("Kate",axis=0)
```

```python
data
```

```python
data.mean(axis="index")
```

```python
data.mean(axis="columns")
```
